```python
# Task 2: Sorting Nodes Alphabetically by Type and Listing the First Three Nodes

# Dictionary to store sorted nodes for each node type
sorted_nodes = {}

# Loop through each unique node type found in the dataset
for node_type in pd.concat([kg['x_type'], kg['y_type']]).unique():
    # Filter nodes with the current type from x_type and y_type
    # Extract relevant columns (index and name) for both x and y nodes
    x_nodes = kg[kg['x_type'] == node_type][['x_index', 'x_name']].drop_duplicates()
    y_nodes = kg[kg['y_type'] == node_type][['y_index', 'y_name']].drop_duplicates()
    
    # Rename columns for consistency when merging x and y nodes
    x_nodes = x_nodes.rename(columns={'x_index': 'index', 'x_name': 'name'})
    y_nodes = y_nodes.rename(columns={'y_index': 'index', 'y_name': 'name'})
    
    # Combine x and y nodes into a single DataFrame, removing duplicates
    all_nodes = pd.concat([x_nodes, y_nodes]).drop_duplicates()
    
    # Sort the combined nodes alphabetically by name
    sorted_nodes[node_type] = all_nodes.sort_values('name').head(3)  # Keep only the first 3 nodes

# Display the sorted nodes for each node type
for node_type, nodes in sorted_nodes.items():
    print(f"Node Type: {node_type}")
    print(nodes)
    print()
```

```python
# Task 3: Create Node-Relation Frequency Table

# Initialize an empty DataFrame to store frequency counts
node_relation_frequency = pd.DataFrame()

# Loop through each unique relation type in the dataset
for relation in kg['relation'].unique():
    # Filter rows corresponding to the current relation
    relation_data = kg[kg['relation'] == relation]

    # Count occurrences of each x_type (source node type) for this relation
    x_counts = relation_data['x_type'].value_counts().rename('source_counts')

    # Count occurrences of each y_type (target node type) for this relation
    y_counts = relation_data['y_type'].value_counts().rename('target_counts')

    # Combine x and y counts into a single DataFrame
    relation_counts = pd.concat([x_counts, y_counts], axis=1).fillna(0)  # Fill missing values with 0
    relation_counts['relation'] = relation  # Add a column for the relation type

    # Append this relation's counts to the main frequency DataFrame
    node_relation_frequency = pd.concat([node_relation_frequency, relation_counts])

# Reset the index for a clean table
node_relation_frequency.reset_index(inplace=True)

# Rename columns for clarity
node_relation_frequency.rename(columns={'index': 'node_type'}, inplace=True)

# Display the frequency table
print(node_relation_frequency)
```

```python
# Task 3.1: Heatmap of Node-Relation Frequencies
import seaborn as sns

# Pivot the frequency table for heatmap compatibility
pivot_table = node_relation_frequency.pivot_table(
    index='relation', 
    columns='node_type', 
    values=['source_counts', 'target_counts'], 
    aggfunc='sum', 
    fill_value=0
)

# Plot the heatmap
plt.figure(figsize=(16, 7))
sns.heatmap(pivot_table['source_counts'], annot=True, fmt='g', cmap='coolwarm')
plt.title('Node-Relation Frequency Heatmap (Source Counts)')
plt.ylabel('Relation Types')
plt.xlabel('Node Types')
plt.tight_layout()
plt.show()
```

```python
import networkx as nx

# Task 2.1: Construct the Knowledge Graph
# Initialize an empty directed graph
G = nx.DiGraph()

# Add edges from the dataset
for _, row in kg.iterrows():
    # Use x_name and y_name as node names, relation as edge label
    G.add_edge(row['x_name'], row['y_name'], relation=row['relation'])

# Print basic stats about the graph
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
# Step 2: Initialize a directed graph
knowledge_graph = nx.DiGraph()

# Step 3: Add nodes and edges
for _, row in kg.iterrows():
    # Add nodes with attributes (e.g., type and name)
    knowledge_graph.add_node(row['x_name'], node_type=row['x_type'])
    knowledge_graph.add_node(row['y_name'], node_type=row['y_type'])
    
    # Add edges with relation as an attribute
    knowledge_graph.add_edge(row['x_name'], row['y_name'], relation=row['relation'])

# Step 4: Visualize the knowledge graph
# Sample the graph if it is too large for visualization
if len(knowledge_graph.nodes) > 100:
    sampled_graph = knowledge_graph.subgraph(list(knowledge_graph.nodes)[:100])  # Sample first 100 nodes
else:
    sampled_graph = knowledge_graph

# Define colors for node types
node_types = list({data['node_type'] for _, data in knowledge_graph.nodes(data=True)})
palette = sns.color_palette("Set3", len(node_types))
node_color_map = dict(zip(node_types, palette))

# Get node colors based on their type
node_colors = [node_color_map[knowledge_graph.nodes[node]['node_type']] for node in sampled_graph.nodes]

# Create the layout
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(sampled_graph, seed=42)

# Draw nodes
nx.draw_networkx_nodes(
    sampled_graph,
    pos,
    node_size=800,
    node_color=node_colors,
    alpha=0.9
)

# Draw edges
nx.draw_networkx_edges(
    sampled_graph,
    pos,
    arrows=True,
    arrowstyle='->',
    arrowsize=15,
    edge_color="gray",
    width=1.2
)

# Draw labels for nodes
nx.draw_networkx_labels(
    sampled_graph,
    pos,
    font_size=8,
    font_color="black"
)

# Draw edge labels
edge_labels = nx.get_edge_attributes(sampled_graph, 'relation')
nx.draw_networkx_edge_labels(
    sampled_graph,
    pos,
    edge_labels=edge_labels,
    font_size=7,
    font_color="blue"
)

# Add a legend for node types
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=node_color_map[node], markersize=10, label=node)
    for node in node_types
]
plt.legend(
    handles=legend_elements,
    loc='upper left',
    bbox_to_anchor=(1, 1),
    fontsize=8,
    title="Node Types",
    title_fontsize=10
)

# Add title and display
plt.title("Knowledge Graph", fontsize=16)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from community import community_louvain
import matplotlib.cm as cm

# Step 1: Recreate the ontology graph
ontology_graph = nx.DiGraph()

# Adding edges to ontology graph based on node types and relations
for _, row in kg.iterrows():
    ontology_graph.add_edge(row['x_type'], row['y_type'], relation=row['relation'])

# Step 2: Perform clustering using Louvain community detection
# Convert directed graph to undirected for community detection
undirected_ontology = ontology_graph.to_undirected()
partition = community_louvain.best_partition(undirected_ontology)

# Step 3: Assign colors to clusters
# Get unique community IDs
unique_communities = set(partition.values())
community_colors = {community: color for community, color in zip(unique_communities, cm.tab20.colors)}

# Step 4: Draw the clustered ontology graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(ontology_graph, seed=42)  # For consistent layout

# Draw nodes with cluster-specific colors
for node, community in partition.items():
    nx.draw_networkx_nodes(
        ontology_graph,
        pos,
        nodelist=[node],
        node_color=[community_colors[community]],
        label=f"Community {community}",
        node_size=800,
    )

# Draw edges and labels
nx.draw_networkx_edges(ontology_graph, pos, alpha=0.5)
nx.draw_networkx_labels(ontology_graph, pos, font_size=10, font_color="black")

# Add legend for communities
legend_labels = [f"Community {community}" for community in unique_communities]
plt.legend(legend_labels, loc="best", fontsize="small", frameon=True, title="Clusters")

# Add title
plt.title("Clustered Ontology Visualization")
plt.show()
```

```python
import networkx as nx

# Recreate the graph G with an example dataset for demonstration
G = nx.DiGraph()

# Adding sample nodes and edges to simulate a disease-related subgraph
G.add_node("Disease_A", type="disease")
G.add_node("Disease_B", type="disease")
G.add_node("Protein_X", type="protein")
G.add_node("Drug_Y", type="drug")
G.add_node("Disease_C", type="disease")

# Adding edges
G.add_edge("Disease_A", "Protein_X")
G.add_edge("Protein_X", "Disease_B")
G.add_edge("Protein_X", "Drug_Y")
G.add_edge("Drug_Y", "Disease_C")

# Helper function to find two disease nodes satisfying the conditions
def find_disease_nodes(graph, disease_nodes):
    for i, node1 in enumerate(disease_nodes):
        for node2 in disease_nodes[i + 1:]:
            # Check if nodes are not directly connected
            if not graph.has_edge(node1, node2) and not graph.has_edge(node2, node1):
                try:
                    # Check if they are separated by no more than three edges
                    if nx.shortest_path_length(graph, source=node1, target=node2) <= 3:
                        return node1, node2
                except nx.NetworkXNoPath:
                    continue
    return None, None

# Get all disease nodes
disease_nodes = [node for node, attr in G.nodes(data=True) if attr.get('type') == 'disease']

# Find two disease nodes that satisfy the conditions
node_1, node_2 = find_disease_nodes(G, disease_nodes)

node_1, node_2
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Manually define a small static graph
static_edges = [
    ("Disease1", "Disease2"),
    ("Disease2", "Gene1"),
    ("Gene1", "Drug1"),
    ("Drug1", "Disease3"),
    ("Disease3", "Anatomy1"),
]

# Create a new graph with static edges
static_graph = nx.DiGraph()
static_graph.add_edges_from(static_edges)

# Visualize the static graph
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(static_graph, seed=42)
nx.draw(
    static_graph,
    pos,
    with_labels=True,
    node_size=500,
    node_color="lightgreen",
    edge_color="black",
    font_size=10
)
plt.title("Tiny Static Subgraph for Visualization")
plt.show()
```

```python
# Step 1: Node Degree Analysis
# Calculate in-degree, out-degree, and total degree
in_degrees = dict(G.in_degree())  # Incoming edges
out_degrees = dict(G.out_degree())  # Outgoing edges
total_degrees = {node: in_degrees[node] + out_degrees[node] for node in G.nodes()}

# Find the top 10 nodes by total degree
top_nodes_by_degree = sorted(total_degrees.items(), key=lambda x: x[1], reverse=True)[:10]

# Print results
print("Top 10 Nodes by Total Degree:")
for node, degree in top_nodes_by_degree:
    print(f"Node: {node}, Total Degree: {degree}")
```

```python
# Step 2: Centrality Analysis (Focus on PageRank)

# Calculate PageRank
pagerank = nx.pagerank(G)
top_pagerank = sorted(pagerank.items(), key=lambda x: x[1], reverse=True)[:10]

# Print the top nodes by PageRank
print("\nTop 10 Nodes by PageRank:")
for node, rank in top_pagerank:
    print(f"Node: {node}, PageRank: {rank:.6f}")
```

```python
# Task 4.1: Community-Specific Statistics
import pandas as pd

# Initialize a dictionary to store statistics
community_stats = {}

for community_id in set(partition.values()):
    # Get nodes belonging to this community
    community_nodes = [node for node, comm_id in partition.items() if comm_id == community_id]
    subgraph = undirected_graph.subgraph(community_nodes)  # Subgraph for this community
    
    # Compute statistics
    num_nodes = len(community_nodes)
    num_edges = subgraph.number_of_edges()
    avg_degree = sum(dict(subgraph.degree()).values()) / num_nodes
    top_nodes = sorted(subgraph.degree, key=lambda x: x[1], reverse=True)[:3]  # Top 3 nodes by degree
    
    community_stats[community_id] = {
        "Number of Nodes": num_nodes,
        "Number of Edges": num_edges,
        "Average Degree": avg_degree,
        "Top Nodes": [node for node, _ in top_nodes]
    }

# Convert statistics to a DataFrame for better readability
community_stats_df = pd.DataFrame.from_dict(community_stats, orient="index")
community_stats_df.index.name = "Community ID"

# Display the community statistics
print(community_stats_df)
```

```python
# Task 4.2: Inter-Community Relationships
inter_community_edges = []

# Loop through edges in the original graph
for u, v in G.edges():
    community_u = partition.get(u)
    community_v = partition.get(v)
    if community_u != community_v:  # Edge between different communities
        inter_community_edges.append((community_u, community_v))

# Count inter-community edges
inter_community_counts = pd.DataFrame(inter_community_edges, columns=["Community U", "Community V"])
inter_community_counts = inter_community_counts.value_counts().reset_index(name="Edge Count")

# Display inter-community relationship counts
print(inter_community_counts)
```

```python
# Create a DataFrame for inter-community relationships
edges = []
for u, v, data in G.edges(data=True):
    if partition[u] != partition[v]:  # Only consider edges between different communities
        edges.append({
            "Community U": partition[u],
            "Community V": partition[v],
            "Edge Count": 1
        })

# Aggregate edges to calculate total weights between communities
import pandas as pd
inter_community_relationships = pd.DataFrame(edges).groupby(
    ["Community U", "Community V"], as_index=False
).sum()

# Ensure symmetric edges for undirected graph (both directions are counted)
inter_community_relationships = pd.concat([
    inter_community_relationships,
    inter_community_relationships.rename(
        columns={"Community U": "Community V", "Community V": "Community U"}
    )
]).drop_duplicates()

# Display the table for verification
print(inter_community_relationships)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the inter-community relationship graph
community_graph = nx.Graph()

# Add edges with weights (from inter_community_relationships)
for _, row in inter_community_relationships.iterrows():
    u, v, weight = row["Community U"], row["Community V"], row["Edge Count"]
    community_graph.add_edge(u, v, weight=weight)

# Get edge weights for scaling widths
edges = community_graph.edges(data=True)
edge_weights = [data["weight"] for _, _, data in edges]

# Normalize edge weights for better visualization
max_weight = max(edge_weights)
scaled_weights = [weight / max_weight * 5 for weight in edge_weights]  # Scale weights to a max width of 5

# Draw the graph
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(community_graph, seed=42)  # Use a spring layout

# Draw nodes
nx.draw_networkx_nodes(
    community_graph,
    pos,
    node_size=300,
    node_color=list(community_graph.nodes()),
    cmap=plt.cm.tab20,
    alpha=0.8,
)

# Draw edges with scaled widths
nx.draw_networkx_edges(
    community_graph,
    pos,
    alpha=0.5,
    width=scaled_weights,
)

# Draw labels
nx.draw_networkx_labels(
    community_graph,
    pos,
    font_size=10,
    font_color="black",
)

# Add title and display
plt.title("Enhanced Inter-Community Relationship Graph", fontsize=14)
plt.axis("off")
plt.show()
```

```python
# Use BFS to infer new relations in the subgraph
from collections import defaultdict, deque

# Define BFS function
def infer_new_relations(graph, max_depth=6):
    inferred_relations = []
    for start_node in graph.nodes():
        visited = set()
        queue = deque([(start_node, [start_node])])  # Node, Path
        while queue:
            current_node, path = queue.popleft()
            if len(path) > max_depth:
                continue
            for neighbor in graph.neighbors(current_node):
                if neighbor not in visited:
                    new_path = path + [neighbor]
                    inferred_relations.append({
                        "Start": start_node,
                        "End": neighbor,
                        "Path": new_path
                    })
                    queue.append((neighbor, new_path))
                    visited.add(neighbor)
    return inferred_relations

# Run BFS on the subgraph
inferred_relations = infer_new_relations(subgraph, max_depth=6)

# Convert results to a DataFrame for better readability
inferred_relations_df = pd.DataFrame(inferred_relations)

# Select three sample inferred relations to map their reasoning chains
sample_inferences = inferred_relations_df.sample(3)

print("Sample Inferred Relations:")
print(sample_inferences)

# output for further discussion in the report.
```

```python
import pandas as pd

# Load the data files
disease_data = pd.read_csv('disease_features.tab', sep='\t', dtype=str)
drug_data = pd.read_csv('drug_features.tab', sep='\t', dtype=str)

# Step 1: Preprocess and filter
disease_data['group_name_bert'] = disease_data['group_name_bert'].str.strip().str.lower()
drug_data['description'] = drug_data['description'].str.strip().str.lower()

# Filter for the selected drug before merging
selected_drug = "estradiol valerate"
filtered_drug_data = drug_data[drug_data['description'].str.contains(selected_drug, na=False)]

# Step 2: Merge filtered data
merged_data = pd.merge(
    disease_data,
    filtered_drug_data,
    left_on='group_name_bert',
    right_on='description',
    how='inner'
)

# Step 3: Compute joint counts of anatomy and drugs
anatomy_drug_counts = merged_data.groupby(['group_name_bert', 'description']).size().reset_index(name='joint_count')

# Step 4: Compute the count of each drug
drug_counts = filtered_drug_data['description'].value_counts().reset_index(name='drug_count')
drug_counts.rename(columns={'index': 'description'}, inplace=True)

# Step 5: Compute conditional probabilities
anatomy_drug_counts = anatomy_drug_counts.merge(drug_counts, on='description', how='left')
anatomy_drug_counts['P(Anatomy|Drug)'] = anatomy_drug_counts['joint_count'] / anatomy_drug_counts['drug_count']

# Step 6: Filter and display results for the selected drug
conditional_probabilities = anatomy_drug_counts[anatomy_drug_counts['description'] == selected_drug]

# Step 7: Display results
if not conditional_probabilities.empty:
    print(f"Conditional Probabilities (P(Anatomy | Drug)) for {selected_drug}:")
    print(conditional_probabilities[['group_name_bert', 'description', 'P(Anatomy|Drug)']])
else:
    print(f"No conditional probabilities found for the drug: {selected_drug}")

# Step 8: Save results for reporting
output_file = "conditional_probabilities.csv"
conditional_probabilities.to_csv(output_file, index=False)
print(f"Results saved to {output_file}.")
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Define the Bayesian Network structure
model = BayesianNetwork([
    ('Anatomy', 'Protein'),
    ('Protein', 'Disease'),
    ('Protein', 'Drug'),
    ('Disease', 'Drug')
])

# Define CPDs
cpd_anatomy = TabularCPD(variable='Anatomy', variable_card=2,
                         values=[[0.7], [0.3]])  # Example probabilities

cpd_protein_given_anatomy = TabularCPD(variable='Protein', variable_card=3,
                                       values=[[0.6, 0.3],
                                               [0.3, 0.4],
                                               [0.1, 0.3]],
                                       evidence=['Anatomy'],
                                       evidence_card=[2])

cpd_disease_given_protein = TabularCPD(variable='Disease', variable_card=2,
                                       values=[[0.8, 0.4, 0.3],
                                               [0.2, 0.6, 0.7]],
                                       evidence=['Protein'],
                                       evidence_card=[3])

# Updated CPD: Drug depends on both Disease and Protein
cpd_drug_given_disease_and_protein = TabularCPD(variable='Drug', variable_card=2,
                                                values=[
                                                    [0.9, 0.5, 0.6, 0.3, 0.7, 0.2],
                                                    [0.1, 0.5, 0.4, 0.7, 0.3, 0.8]],
                                                evidence=['Disease', 'Protein'],
                                                evidence_card=[2, 3])

# Add CPDs to the model
model.add_cpds(cpd_anatomy, cpd_protein_given_anatomy,
               cpd_disease_given_protein, cpd_drug_given_disease_and_protein)

# Validate the Bayesian Network
try:
    assert model.check_model()
    print("The Bayesian Network is valid.")
except ValueError as e:
    print(f"Model validation error: {e}")

# Print the network structure
print("Bayesian Network Structure:")
print(model.edges())
```

```python
# Perform inference
inference = VariableElimination(model)

# Query 1: Probability of Anatomy given a Drug
query_result = inference.query(variables=['Anatomy'], evidence={'Drug': 1})
print("Probability of Anatomy given Drug 1:")
print(query_result)

# Query 2: Most probable disease given a drug
query_result = inference.map_query(variables=['Disease'], evidence={'Drug': 1})
print("Most probable disease given Drug 1:")
print(query_result)

# Query 3: Joint probability of Drug and Anatomy
query_result = inference.query(variables=['Drug', 'Anatomy'])
print("Joint probability of Drug and Anatomy:")
print(query_result)
```
